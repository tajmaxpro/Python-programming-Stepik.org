```python
def factorial(a, b):
    multiplier=1
    total=0
    for i in range(a,b):
        for j in range(2,i+1):
            multiplier*=j
            total+=multiplier
    print(total)
factorial(3,7)
#factorial(int(input()), int(input())
```

# Модули.
    Таким образом, модуль – это любой файл, содержащий код на языке Python. Для Python термины “библиотека” и “модуль” равнозначны, в других языках программирования библиотекой может называться совокупность нескольких модулей.

Приведем пример использования модульного программирования:
	Допустим, мы написали ИИ, который за нас играет на трех биржах X, Y, Z. Скрипт отличается для каждой из бирж, поэтому каждый алгоритм вынесем в отдельную программу (модуль), а затем создадим файл main.py, который бы “собирал” все модули и запускал каждый скрипт. Тогда наш проект будет состоять из четырех файлов:
stock_X.py – алгоритм для игры на бирже X
stock_Y.py – алгоритм для игры на бирже Y
stock_Z.py – алгоритм для игры на бирже Z
main.py – основная программа. Она импортирует модули и может пользоваться их содержимым. Открыв проект, именно ее мы запустим, чтобы начать исполнение.

# Подключение и создание модуля
	Как уже было сказано, модулем является любая программа на языке Python. Но пока несколько программ не связаны программистом, они могут действовать только отдельно, используя только свои ресурсы. Создадим программу module.py:

def greet_user():
    print("Hello! I'm a brand-new module :3")
text = "Это очень важная строка, не потеряйте ее"

Этот модуль включает в себя всего одну функцию и одну переменную. Создадим файл main.py. Чтобы подключить созданный модуль, нужно использовать команду import <имя файла без расширения .py>. Чтобы использовать содержимое модуля, нужно обращаться к переменной/функции так: <имя модуля без расширения>.<имя объекта> – это так называемая “точечная нотация”.

import module
text = "\nСпасибо, что не потеряли"
print(module.text, text)  # переменные имеют одинаковое имя, но из-за точки проблем не возникает
module.greet_user()

Важно отметить, что Python будет искать модули, перечисленные после import, сначала в текущей папке, затем в папке, в которой установлен Python. Поэтому модули нужно создавать в той же папке, где планируется создать основную программу. 
Несколько модулей можно импортировать через запятую:

import module1, module2, ..., module_n

	Часто необходимо загружать не весь модуль, а только некоторые его компоненты. Если из большого модуля нам нужны несколько функций, нет необходимости загружать его целиком. Чтобы выборочно импортировать объекты модуля используют конструкцию

from <имя модуля> import <компонент1, компонент2, ..., компонентX>
	
    Доступ к объектам, импортированным таким образом, можно получить без точки. Например, мы могли бы написать в файле main.py:

from module import greet_user  # заметьте, что, импортируя функцию, мы не пишем круглые скобки
greet_user()

	В конструкции from … import можно импортировать и все объекты модуля:

from <имя модуля> import *
	
    Это позволит обращаться ко всем объектам модуля без использования точки. Однако это используется редко, так как может привести к проблемам. Например, если бы мы в первой версии программы main.py написали в начале from module import *, то у нас вместо двух переменных module.text и text оказались бы переменные text и text. Что произойдет, если попытаться напечатать их?
	Если названия модуля слишком длинное, то обращение ко всем объектам через точку может серьезно ухудшить читаемость или тратить Ваше время. В таких случаях при импортировании модуля можно указать его псевдоним с помощью слова as. Изменим main.py:
import module as m

text = "\nСпасибо, что не потеряли"
print(m.text, text)
m.greet_user()

	Создав переменную image, мы можем узнать некоторые параметры изображения – для обработки нам понадобятся размер изображения и массив пикселей. 
Справка: компьютер представляет изображение в виде матрицы пикселей, т. е. если размер изображения 600 * 400, то в матрице 400 строк и 600 столбцов. Каждый пиксель имеет свой цвет, так как он кодируется четырьмя числами от 0 до 255 – компонентами красного, зеленого и синего цветов (RGB), а также яркостью (A). Собирая изображение из тысяч пикселей разных цветов, компьютер отображает понятную человеку картинку. Таким образом, картинка – это матрица кортежей длины 3, или иначе – трехмерный массив.


```python
#Получим параметры изображения:
width = image.width  # определяем ширину
height = image.height  # определяем высоту
pixels = image.load()  # выгружаем значения пикселей
```

```python
draw = ImageDraw.Draw(image)
def get_new_color(color):
    new_color = color - randint(50, 100)  # для начала просто отнимем случайное число
    return max(new_color, 0)  # исключаем ситуацию, когда новый цвет стал меньше нуля
for i in range(width):
    for j in range(height):
        colors = pixels[i, j]  # получим кортеж (RGBA)
        red = colors[0]
        green = colors[1]
        blue = colors[2]
        new_red = get_new_color(red)
        new_green = get_new_color(green)
        new_blue = get_new_color(blue)
        # перерисуем в точке (i, j) пиксель с новыми компонентами (RGB)
        draw.point((i, j), (new_red, new_green, new_blue))
image.save("test.png")
```

# Способы задания массивов, их копирование
    Как ранее говорилось, одним из важных отличий numpy-массивов от обычных списков является то, что все элементы массива имеют одинаковый тип данных. Основные типы данных в массивах: 

bool (логический тип)

int (целые со знаком)

float (действительные)

complex (комплексные, с этим множеством чисел вы познакомитесь в старших классах )

str (строки)

Мы не будем рассматривать все возможные типы данных (np.sctypes), так как от перечисленных выше они отличаются только размером в байтах (например,  int64, float32).
Создание массива можно выполнить различными способами. Первый из них (рассматривался на прошлом уроке) – преобразовать обычный Python-список или кортеж в массив:

import numpy as np

myArray1 = np.array([[1, 2], [3, 4]]) # array([[1, 2], [3, 4]])

myArray2 = np.array(((5, 6), (7, 8))) # array([[5, 6], [7, 8]])

myArray3 = np.array([1, 2, 3], dtype='float') # array([1., 2., 3.])

В данном примере особо интересен массив myArray3. Мы явно преобразовали изначально целочисленный список в массив с вещественными числами, поэтому после чисел теперь стоят точки.
Подумайте самостоятельно, а затем проверьте, какой вид будет иметь массив myArray4.

	Также можно создать массив указанного размера, заполненный нулями или единицами. В линейной алгебре часто встречаются единичные матрицы (в общем и целом, в нашем случае это двумерный массив), можно сказать, что это квадратная матрица из нулей, в которой на главной диагонали расположены единицы. 
  	Рассмотрим следующие примеры:

zerosArray = np.zeros((2, 2)) # array([[0., 0.], [0., 0.]]) 

onesArray = np.ones((3, 1)) # array([[1.],[1.],[1.]])

identityArray = np.eye(2) # array([[1., 0.], [0., 1.]]) ⇐ Единичная матрица

Подумайте и ответьте на вопрос: почему в последнем примере функция np.eye() принимает не кортеж, а одно целое число?
